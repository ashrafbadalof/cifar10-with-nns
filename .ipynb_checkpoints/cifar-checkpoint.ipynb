```python
import numpy as np
import tensorflow as tf
import keras
from keras import layers
```

```python
cifar = tf.keras.datasets.cifar10
(X_train_full, y_train_full), (X_test, y_test) = cifar.load_data()
```

```python
y_train_flat = y_train_full.flatten()
y_train_flat.shape
```

```python
class_names  = {
    0: "airplane", 
    1: "automobile", 
    2: "bird", 
    3: "cat", 
    4: "deer", 
    5: "dog", 
    6: "frog", 
    7: "horse", 
    8: "ship", 
    9: "truck"
}

y_train_repr = [class_names[label] for label in y_train_flat]
np.unique(y_train_repr)
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,8))
for i in range(36):
    plt.subplot(6,6,i+1)
    plt.imshow(X_train_full[i])
    plt.title(y_train_repr[i])
    plt.axis("off")
plt.tight_layout()
```

```python
X_train, y_train = X_train_full[:45000] / 250.0, y_train_full[:45000]
X_valid, y_valid = X_train_full[45000:] / 250.0, y_train_full[45000:]
X_test = X_test / 250.0
```

```python
print(X_train.shape)
print(X_valid.shape)
```

```python
X_test[0]
```

```python
def create_model(neurons1, neurons2, neurons3, neurons4, neurons5):
    model = keras.Sequential([
    keras.layers.Flatten(input_shape = [32, 32, 3]),
    keras.layers.Dropout(0.1),    
    keras.layers.Dense(neurons1, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(neurons2, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(neurons3, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(neurons4, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(neurons5, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),    
    keras.layers.Dense(10, activation = 'softmax'),
])
    model.compile(
    loss = 'sparse_categorical_crossentropy',
    optimizer = 'adam',
    metrics = ["accuracy"],
)
    return model

def build_model(hp):
    neurons1 = hp.Int("1", min_value = 32, max_value = 512, step = 32)
    neurons2 = hp.Int("2", min_value = 32, max_value = 512, step = 32)
    neurons3 = hp.Int("3", min_value = 32, max_value = 512, step = 32)
    neurons4 = hp.Int("4", min_value = 32, max_value = 512, step = 32)
    neurons5 = hp.Int("5", min_value = 32, max_value = 512, step = 32)

    model = create_model(
        neurons1 = neurons1, neurons2 = neurons2, neurons3 = neurons3, neurons4 = neurons4, neurons5 = neurons5
    )
    return model
```

```python
import keras_tuner as kt

tuner = kt.BayesianOptimization(
    build_model,
    objective = "val_loss",
    max_trials = 20,
    overwrite = True
)
```

```python
tuner.search(X_train, y_train, validation_data = (X_valid, y_valid), epochs = 20, batch_size = 32)
```

```python
best_model = tuner.get_best_models()[0]
```

```python
best_hyperparams = tuner.get_best_hyperparameters(num_trials=1)[0]

print(best_hyperparams.get("1"))
print(best_hyperparams.get("2"))
print(best_hyperparams.get("3"))
print(best_hyperparams.get("4"))
print(best_hyperparams.get("5"))
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = [32, 32, 3]),
    keras.layers.Dropout(0.1),    
    keras.layers.Dense(512, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(512, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(512, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(512, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),
    keras.layers.Dense(512, kernel_initializer = 'he_normal', activation = 'relu'),
    keras.layers.Dropout(0.1),    
    keras.layers.Dense(10, activation = 'softmax'),
])
```

```python
history = best_model.fit(X_train, y_train, validation_data = (X_valid, y_valid), epochs = 20, batch_size = 32)
```

```python
import pandas as pd

plt.figure(figsize = (10,8))
dataframe = pd.DataFrame(history.history)
dataframe.loc[:, ['loss', 'val_loss']].plot();
```
